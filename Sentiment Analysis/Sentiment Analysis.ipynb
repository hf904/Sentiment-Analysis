```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer()
```

```python
#Reading in data

data=pd.read_csv("Train.csv")

data.head()
```

```python
#Plotting Piechart 

fig= plt.figure(figsize=(5,5))
colors= ["skyblue","pink"]
pos= data[data["label"]==1]
neg= data[data["label"]==0]
ck= [pos["label"].count(),neg["label"].count()]
legpie= plt.pie(ck,labels=["Positive","Negative"],
               autopct="%1.1f%%",
               shadow = True,
               colors = colors,
               startangle = 45,
               explode = (0, 0.1))
```

```python
#Performing Data Preprocessing

import re
def preprocessor(text):
             text=re.sub('<[^>]*>','',text)
             emojis=re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)',text)
             text=re.sub('[\W]+',' ',text.lower()) +\
                ' '.join(emojis).replace('-','')
             return text   
data['text']=data['text'].apply(preprocessor)
```

```python
#Applying Stemmer

from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()

def tokenize(text):
    return text.split()

def tokenizer_porter(text):
    return [porter.stem(word) for word in text.split()]
    
```

```python
#Removing stopwords

import nltk
nltk.download("stopwords")

from nltk.corpus import stopwords
stop = stopwords.words("english")

```

```python
# Using wordcloud to visualize Positive & Negative Sentiments

from wordcloud import WordCloud

positivedata = data[data["label"]==1]
positivedata = positivedata["text"]
negativedata = data[data["label"]==0]
negativedata = negativedata["text"]


def wordcloud_draw(data, color="white"):
    words = " ".join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if(word!= "movie" and word!= "film")
                            ])
    wordcloud = WordCloud(stopwords = stop,
                         background_color = color,
                         width = 2500,
                         height = 2000
                         ).generate(cleaned_word)
    plt.figure(1,figsize=(10,7))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()
    
    print("Positive Sentiments:")
    wordcloud_draw(positivedata, "white")
    print("Negative Sentiments:")
    wordcloud_draw(negativedata)
```

```python
from wordcloud import WordCloud
positivedata = data[ data['label'] == 1]
positivedata =positivedata['text']
negdata = data[data['label'] == 0]
negdata= negdata['text']

def wordcloud_draw(data, color = 'white'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                              if(word!='movie' and word!='film')
                            ])
    wordcloud = WordCloud(stopwords=stop,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(10, 7))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
print("Positive Sentiments")
wordcloud_draw(positivedata,'white')
print("Negative Sentiments")
wordcloud_draw(negdata)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(strip_accents=None,lowercase=False,preprocessor=None,tokenizer=tokenizer_porter,use_idf=True,norm='l2',smooth_idf=True)
y=data.label.values
x=tfidf.fit_transform(data.text)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state=1, test_size=0.5, shuffle=False)
```

```python
from sklearn.linear_model import LogisticRegressionCV
clf = LogisticRegressionCV(cv = 6, scoring="accuracy", random_state=0, n_jobs=-1, max_iter=500).fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
